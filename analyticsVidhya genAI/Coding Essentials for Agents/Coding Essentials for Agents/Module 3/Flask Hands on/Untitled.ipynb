```python
import requests

# URL of the Flask endpoint
url = 'http://127.0.0.1:5050/submit'

# Data to be sent in the POST request
data = {
    'message': 'Hello, Flask!'
}

# Make the POST request
response = requests.post(url, data=data)

# Print the response content
print(response.text)
```
